```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('iris_csv.csv')
plt.figure(figsize=(10,5))
x=data["sepallength"]
plt.hist(x,bins=20,color="red")
plt.title("Sepal length in cm")
plt.xlabel("Sepal length in cm")
plt.ylabel("count")
plt.show()
```
